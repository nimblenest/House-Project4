```python
import pandas as pd
from datetime import datetime
import numpy as np
```

## Import and concat CSV Files

```python
#relative paths to csv files
df1 = pd.read_csv('Files/2019 Sales 1-6.csv')
df2 = pd.read_csv('Files/2019 Sales 7-12.csv')
df3 = pd.read_csv('Files/2020 Sales 1-12.csv')
df4 = pd.read_csv('Files/2021 Sales 1-12.csv')
df5 = pd.read_csv('Files/2022 Sales 1-12.csv')
df6 = pd.read_csv('Files/2023 Sales 1-12.csv')

#combine files into a single df
combined_sales_df = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)

combined_sales_df
```

## Handling null values and duplicates

```python
combined_sales_df.isnull().sum()
```

```python
# drop unnecessary columns
combined_sales_df.drop(['Pics', 'Status', 'Address', 'Subdivision Full', 'Price Per SQFT', 'Sold Price Per SQFT', 'Sold Price/Sqft', 'Waterfront Access', 'Zoning', 'Zip'], axis=1, inplace=True)
combined_sales_df
```

```python
# drop rows with missing values
combined_sales_df.dropna(subset=['YrBlt', 'Beds', 'FBths', 'HBths', 'Sold Price', 'Approx Lot SqFt', '# of Rooms'], inplace=True)
combined_sales_df
```

```python
# check for duplicate rows
combined_sales_df.duplicated().sum()
```

```python
# drop duplicate rows 
combined_sales_df.drop_duplicates(inplace=True)
```

## Data types and column cleanup

```python
# get the current year
current_year = datetime.now().year

# add a dynamic "age of house" column
combined_sales_df['Age of House'] = current_year - combined_sales_df['YrBlt']
```

```python
# look at datatypes
combined_sales_df.info()
```

```python
# remove dollar signs and commas and convert to numeric
for col in ['LvngAreaSF', 'List Price', 'Sold Price', 'Approx Lot SqFt']:
    combined_sales_df[col] = combined_sales_df[col].str.replace(',', '').str.replace('$', '')
    combined_sales_df[col] = pd.to_numeric(combined_sales_df[col], errors='coerce')

# convert to numeric
combined_sales_df['# of Rooms'] = pd.to_numeric(combined_sales_df['# of Rooms'], errors='coerce')
combined_sales_df['Age of House'] = pd.to_numeric(combined_sales_df['Age of House'], errors='coerce')

#convert closing date to datetime
combined_sales_df['Closing Date'] = pd.to_datetime(combined_sales_df['Closing Date'])
```

```python
# Basement column cleanup

#fill nulls with 0, convert to boolean
combined_sales_df['Basement'] = combined_sales_df['Basement'].replace({'No': 0, 'Yes': 1})
combined_sales_df['Basement'].fillna(0, inplace=True)
combined_sales_df['Basement'] = combined_sales_df['Basement'].astype(bool)
```

```python
# Garage column cleanup

# fill null values with 0
combined_sales_df['Garage'].fillna(0, inplace=True)

# replace any value of 1 or higher with 1
combined_sales_df['Garage'] = combined_sales_df['Garage'].apply(lambda x: 1 if x >= 1 else 0)

# convert to boolean
combined_sales_df['Garage'] = combined_sales_df['Garage'].astype(bool)
```

```python
# convert closing date to datetime
combined_sales_df['Closing Date'] = pd.to_datetime(combined_sales_df['Closing Date'])
```

```python
columns_to_convert = ['YrBlt', 'Beds', 'FBths', 'HBths', '# of Rooms', 'Approx Lot SqFt', 'Total Baths', 'Age of House']

for col in columns_to_convert:
    combined_sales_df[col] = combined_sales_df[col].astype('int64', errors='ignore')
```

```python
combined_sales_df.info()
```

```python
# create a new column calculating price/sqft
combined_sales_df['Price/SqFt'] = combined_sales_df['Sold Price'] / combined_sales_df['LvngAreaSF']
```

```python
combined_sales_df.info()
```

```python
# handle inf values in price/sqft
combined_sales_df['Price/SqFt'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# treat city/county as proper nouns for consistency
combined_sales_df.loc[:, 'City'] = combined_sales_df['City'].str.title()
combined_sales_df.loc[:, 'County'] = combined_sales_df['County'].str.title()

```

```python
# eliminate rows where the list price and sold price are drastically different (potential outliers)

# create a mask for sales where the sold price is more than $50,000 less than the list price
mask = combined_sales_df['Sold Price'] >= (combined_sales_df['List Price'] - 50000)

# apply the mask to the DataFrame to filter out the rows that do not meet the condition
combined_sales_df = combined_sales_df[mask]
```

```python
# look at unique waterfront types
unique_waterfront_types = combined_sales_df['Waterfront Type'].unique()
print(unique_waterfront_types)
```

```python
# convert 'Waterfront Type' into a boolean columnn where lake and river are true and everything else is false
combined_sales_df.loc[:, 'Waterfront Type'] = combined_sales_df['Waterfront Type'].isin(['Lake', 'River']).astype(bool)
```

```python
combined_sales_df.info()
```

```python
# making an explicit copy to avoid SettingWithCopyWarning
combined_sales_df = combined_sales_df.copy()
```

```python
# rename columns to reflect their boolean nature
combined_sales_df.rename(columns={
    'Garage': 'Has Garage',
    'Basement': 'Has Basement',
    'Waterfront Type': 'Is Lake/River Waterfront'
}, inplace=True)
```

```python
combined_sales_df
```

```python
# create calculations for selling vs list price difference and ratio
combined_sales_df['Price Difference'] = combined_sales_df['List Price'] - combined_sales_df['Sold Price']
combined_sales_df['Price Ratio'] = combined_sales_df['Sold Price'] / combined_sales_df['List Price']
```

```python
# drop out of area county values
combined_sales_df = combined_sales_df.loc[combined_sales_df['County'] != 'Out Of Area']
```

## Handling outliers

```python
print(combined_sales_df.describe())
```

```python
# set limits for year built column
combined_sales_df = combined_sales_df[(combined_sales_df['YrBlt'] >= 1800) & (combined_sales_df['YrBlt'] <= current_year)]

# recalculate 'Age of House'
combined_sales_df['Age of House'] = current_year - combined_sales_df['YrBlt']
```

```python
# filter DataFrame for bedrooms and bathrooms within a reasonable range
filtered_df = combined_sales_df[
    (combined_sales_df['Beds'] >= 1) & (combined_sales_df['Beds'] <= 10) &
    (combined_sales_df['FBths'] >= 1) &
    ((combined_sales_df['FBths'] + combined_sales_df['HBths']) <= 10)
]

# assign the filtered DataFrame back to combined_sales_df
combined_sales_df = filtered_df.copy()

# reset the index after filtering
combined_sales_df.reset_index(drop=True, inplace=True)
```

```python
# eliminate unreasonable total # of rooms
combined_sales_df = combined_sales_df[combined_sales_df['# of Rooms'] >= 2]
combined_sales_df = combined_sales_df[combined_sales_df['# of Rooms'] <= 20]
```

```python
print(combined_sales_df.describe())
```

```python
# handle lot sqft outliers
max_lot_sqft_threshold = 1e6  # 1,000,000 sqft threshold for maximum lot size

combined_sales_df = combined_sales_df[
    (combined_sales_df['LvngAreaSF'] > 50) &
    (combined_sales_df['Approx Lot SqFt'] > 50) &
    (combined_sales_df['Approx Lot SqFt'] < max_lot_sqft_threshold)
]
```

```python
print(combined_sales_df.describe())
```

```python
# eliminate outliers in price/sqft

# calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = combined_sales_df['Price/SqFt'].quantile(0.25)
Q3 = combined_sales_df['Price/SqFt'].quantile(0.75)

# calculate IQR
IQR = Q3 - Q1

# define upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# identify outliers
outliers = combined_sales_df[(combined_sales_df['Price/SqFt'] < lower_bound) | (combined_sales_df['Price/SqFt'] > upper_bound)]

# print the outliers
print("Outliers:")
print(outliers)

# remove outliers from the DataFrame
combined_sales_df = combined_sales_df[(combined_sales_df['Price/SqFt'] >= lower_bound) & (combined_sales_df['Price/SqFt'] <= upper_bound)]
```

```python
print(combined_sales_df.describe())
```

```python
# eliminate outliers in list price and sold price

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for List Price
Q1_list_price = combined_sales_df['List Price'].quantile(0.25)
Q3_list_price = combined_sales_df['List Price'].quantile(0.75)

# Calculate IQR for List Price
IQR_list_price = Q3_list_price - Q1_list_price

# Define upper and lower bounds for outliers for List Price
upper_bound_list_price = Q3_list_price + 1.5 * IQR_list_price
lower_bound_list_price = Q1_list_price - 1.5 * IQR_list_price

# Identify outliers for List Price
outliers_list_price = combined_sales_df[(combined_sales_df['List Price'] < lower_bound_list_price) | (combined_sales_df['List Price'] > upper_bound_list_price)]

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for Sold Price
Q1_sold_price = combined_sales_df['Sold Price'].quantile(0.25)
Q3_sold_price = combined_sales_df['Sold Price'].quantile(0.75)

# Calculate IQR for Sold Price
IQR_sold_price = Q3_sold_price - Q1_sold_price

# Define upper and lower bounds for outliers for Sold Price
upper_bound_sold_price = Q3_sold_price + 1.5 * IQR_sold_price
lower_bound_sold_price = Q1_sold_price - 1.5 * IQR_sold_price

# Identify outliers for Sold Price
outliers_sold_price = combined_sales_df[(combined_sales_df['Sold Price'] < lower_bound_sold_price) | (combined_sales_df['Sold Price'] > upper_bound_sold_price)]

# Print the outliers for List Price and Sold Price
print("Outliers for List Price:")
print(outliers_list_price)
print("\nOutliers for Sold Price:")
print(outliers_sold_price)

# Remove outliers from the DataFrame for List Price and Sold Price
combined_sales_df = combined_sales_df[(combined_sales_df['List Price'] >= lower_bound_list_price) & (combined_sales_df['List Price'] <= upper_bound_list_price)]
combined_sales_df = combined_sales_df[(combined_sales_df['Sold Price'] >= lower_bound_sold_price) & (combined_sales_df['Sold Price'] <= upper_bound_sold_price)]
```

```python
print(combined_sales_df.describe())
```

```python
# drop the current 'Total Baths' column
combined_sales_df.drop(columns=['Total Baths'], inplace=True)

# calculate the new total baths column
combined_sales_df['Total Baths'] = combined_sales_df['FBths'] + 0.5 * combined_sales_df['HBths']
```
